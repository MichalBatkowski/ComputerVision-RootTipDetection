```python
# Import necessary libraries
import gym
import numpy as np
import matplotlib.pyplot as plt
import math
import time
from ot2_wrapper_PID import OT2Env
import PID

# Initialize the Gym Pendulum environment
env = OT2Env(render=True)

def generate_setpoints(N):
    ''''
    Generate N random setpoints within the workspace limits
    Bounding Box (min, max):
    x: (-0.1871, 0.2532)
    y: (-0.1707, 0.2197)
    z: (0.1195, 0.2896)
    '''

    x_min, x_max = -0.1871, 0.2532
    y_min, y_max = -0.1707, 0.2197
    z_min, z_max = 0.1195, 0.2896
    
    setpoints = np.zeros((N, 3))
    for i in range(N):
        setpoints[i] = [
            np.random.uniform(x_min, x_max),
            np.random.uniform(y_min, y_max),
            np.random.uniform(z_min, z_max)
        ]
    return setpoints

# Generate N setpoints
setpoints = generate_setpoints(5)

# Define simulation parameters
max_steps = 1000  # Number of steps per episode 

# Control loop
for setpoint_x, setpoint_y, setpoint_z in setpoints:

    # Data storage for plotting
    time_history = []
    x_history = []
    y_history = []
    z_history = [] 

    PID_x = PID.PID(70, 1, 5, setpoint=setpoint_x, output_limits=(-1,1))
    PID_y = PID.PID(70, 1, 5, setpoint=setpoint_y, output_limits=(-1,1))
    PID_z = PID.PID(70, 1, 5, setpoint=setpoint_z, output_limits=(-1,1))

    state, _ = env.reset(setpoint_x, setpoint_y, setpoint_z ,seed=42)
    print(state)
    start_time = time.time()
    
    for step in range(max_steps):
        # Extract state components
        x, y, z = state[:3]
        print(state)

        dt = time.time() - start_time
        # Calculate control action
        vel_x = PID_x.compute(x, dt)
        vel_y = PID_y.compute(y, dt)
        vel_z = PID_z.compute(z, dt)

        # Apply action to the environment
        action = [vel_x, vel_y, vel_z, 0]
        state, reward, done, truncated, info = env.step(action)
        
        # Record data
        time_history.append(dt)
        x_history.append(x)
        y_history.append(y)
        z_history.append(z)
        
        # Sleep to match the sample time (optional)
        time.sleep(0.01)
        
        if done or truncated:
            print(f"Finished after {step+1} steps")
            break

    # Plotting the results
    plt.figure(figsize=(15, 10))

    # Plot Theta (Angle)
    plt.subplot(3, 1, 1)
    plt.plot(time_history, x_history, label='X position (m)')
    plt.axhline(y=setpoint_x, color='r', linestyle='--', label='Setpoint')
    plt.title(f"X joint position over time done in {step+1} steps")
    plt.ylabel('Position (m)')
    plt.legend()
    plt.grid(True)

    # Plot Omega (Angular Velocity)
    plt.subplot(3, 1, 2)
    plt.plot(time_history, y_history, label='Y position (m)', color='g')
    plt.axhline(y=setpoint_y, color='r', linestyle='--', label='Setpoint')
    plt.title('Y joint position over time')
    plt.ylabel('Position (m)')
    plt.legend()
    plt.grid(True)

    # Plot Torque
    plt.subplot(3, 1, 3)
    plt.plot(time_history, z_history, label='Z position (m)', color='m')
    plt.axhline(y=setpoint_z, color='r', linestyle='--', label='Setpoint')
    plt.title('Z joint position over time')
    plt.xlabel('Time (s)')
    plt.ylabel('Position (m)')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

env.close()
```
